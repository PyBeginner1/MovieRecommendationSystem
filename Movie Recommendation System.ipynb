```python
import numpy as np
import pandas as pd
import difflib

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data = pd.read_csv('D:/Music/Music/movies.csv')
data.head()
```

```python
data.isna().sum()
```

```python
#Choosing the columns for recommending the movie
data_sel = ['genres', 'keywords', 'tagline', 'cast', 'director']
data_sel
```

```python
for col in data_sel:
    data[col] = data[col].fillna('')
```

```python
data.isnull().sum()
```

```python
#Combining the selected 5 columns
data_combined = data['genres']+ ' ' + data['keywords']+ ' ' + data['tagline']+ ' ' + data['cast']+ ' ' + data['director']
data_combined
```

```python
#Converting text to numerical values
vectorizer = TfidfVectorizer()
```

```python
data_vectors = vectorizer.fit_transform(data_combined)
print(data_vectors)
```

```python
#Getting similarity scores using cosine similarity
similarity = cosine_similarity(data_vectors)
print(similarity)
```

```python
#Getting movie name from user
movie_name = input('Enter movie name: ')
```

```python
#Getting list of all movies from dataset
data_list = data['title'].tolist()
data_list
```

```python
#Finding close match for movie name given by user
find_close_match = difflib.get_close_matches(movie_name, data_list)
find_close_match
```

```python
close_match = find_close_match[0]
close_match
```

```python
#Finding index value of the close_match
index_val = data[data.title == close_match]['index'].values[0]
index_val
```

```python
#Getting list of similar movies
similarity_score = list(enumerate(similarity[index_val]))
similarity_score
```

```python
#Sorting moves based on their score
movie_sort = sorted(similarity_score, key = lambda x: x[1], reverse = True)
movie_sort
```

```python
#Printing suggested movies

print('Movies suggested for you:  \n')

i = 1
for movie in movie_sort:
    index = movie[0]
    title = data[data.index == index]['title'].values[0]
    print(i,'. ',title)
    i = i + 1
    if i >= 25:
        break
```
